В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

# Обычное тематическое событие
    1) DAU(MAU) - Количество уникальных пользователей в день/в месяц.(Можно предположить, что данная метрика должна увеличиться, т.к. появится дополнительный стимул в виде получения новых уникальных предметов)
    
    2) Stickiness rate - Регулярность использования приложения в течение месяца.(Увеличение данной метрики поможет понять заинтересовали ли мы пользователя данным событием)
    
    3) Retention rate - Процент клиентов, которые возвращаются в приложение.
    
    4) ASD(Average Session Duration) - Суммарная продолжительность сессий за период / общее количество сессий за тот же период.
    
    4) CR(Conversion rate) - Конверсия (Поможет оценить влияет ли тематическое событие на привлечение новых пользователей)

# Усложненное тематическое событие
    Можно предположить, что усложнение механики может негативно сказаться на игре. Я бы проанализировал метрики из обычного тематического события + добавил бы к ним:
    
    1) Доля пользователей, которые вышли из игры после неудачной попытки - пользователи не любят откатываться назад, и выполнять задания по новому. Покажет долю пользователей у которых произошло выгорание после неудачи.
    
    2) Среднее количсество откатов - покажет выдержку клиентов ( Для дальнейших тематических событий покажет, усложнять или уменьшать сложность события)
    
    В целом если есть возможность протестировать два события, то необходимо определить отличаются ли стат.значимо пользователи в обеих группах и обратил бы внимание на показатели Retention rate, Stickiness rate, ASD и на фоне их можно сделать вывод о более успешном тематическом событии.
